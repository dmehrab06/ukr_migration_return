```python
a = Value(-4.0)
b = Value(2.0)
c = a + b
d = a * b + b**3
c += c + 1
c += 1 + c + (-a)
d += d * 2 + (b + a).relu()
d += 3 * d + (b - a).relu()
e = c - d
f = e**2
g = f / 2.0
g += 10.0 / f
print(f'{g.data:.4f}') # prints 24.7041, the outcome of this forward pass
g.backward()
print(f'{a.grad:.4f}') # prints 138.8338, i.e. the numerical value of dg/da
print(f'{b.grad:.4f}') # prints 645.5773, i.e. the numerical value of dg/db
```

```python
a = [Value(-4.0),Value(-2.0)]
b = [Value(-2.0),Value(-1.0)]
scale = Value(1.0)
lr = 0.01

for it in range(0,100):
    loss = (a[0]-b[0]*scale)**2+(a[1]-b[1]*scale)**2

    print(f'{loss.data:.4f}') # prints 24.7041, the outcome of this forward pass
    loss.backward()

    print(f'{scale.grad:.4f}')
    
    scale = scale-lr*(scale.grad)
    print('new scale',scale)
```
